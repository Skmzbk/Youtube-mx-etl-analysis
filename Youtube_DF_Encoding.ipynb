```python
import pandas as pd
import json
```

```python
df = pd.read_csv(r"C:\Users\linkm\Documents\Python\Project\Youtube_Trending_Videos\MXvideos.csv",encoding='latin1')
with open(r"C:\Users\linkm\Documents\Python\Project\Youtube_Trending_Videos\MX_category_id.json") as f:
    categories = json.load(f)
```

```python
df
```

```python
df['title', 'channel_title', 'tags', 'description'] = df['title', 'channel_title', 'tags', 'description'].apply(lambda x: x.encode('latin1').decode('utf-8', errors='ignore')if isinstance(x, str) else x)
```

```python
df[df['title'].str.contains('ñ|á|é|í|ó|ú', regex=True)]
```

```python
cat_df = pd.json_normalize(categories['items'])
cat_df = cat_df[['id', 'snippet.title']].rename(columns={'id': 'category_id', 'snippet.title': 'category_name'})
df['category_id'] = df['category_id'].astype(str)

merged = df.merge(cat_df, on='category_id')
```

```python
merged.to_csv(r'C:\Users\linkm\Documents\Python\Project\Youtube_Trending_Videos\Encoded.csv',index = False)
```
