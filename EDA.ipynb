```python
import pandas as pd
import numpy as ny
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\linkm\Documents\Python\Project\Youtube_Trending_Videos\Clean.csv")
```

```python
df.sort_values(by='views', ascending=False).head(10)
```

```python
df.sort_values(by='dislikes', ascending=False).head(10)
```

```python
df.groupby('category_name')[['views','likes','dislikes','comment_count']].sum(numeric_only=True).sort_values(by='views', ascending=False)
```

```python
df.groupby('channel_title')[['views','likes','dislikes','comment_count']].sum(numeric_only=True).sort_values(by='views', ascending=False).head(10)
```

```python
df.groupby('channel_title')[['views','likes','dislikes','comment_count']].sum(numeric_only=True).sort_values(by='dislikes', ascending=False).head(10)
```

```python
df.groupby('channel_title')[['views','likes','dislikes','comment_count']].sum(numeric_only=True).sort_values(by='comment_count', ascending=False).head(10)
```

```python
numeric_cols = ['views', 'likes', 'dislikes', 'comment_count']

corr_matrix = df[numeric_cols].corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='views', y='likes', alpha=0.5)
plt.xscale('log')
plt.yscale('log')
plt.title('Views vs Likes')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='views', y='likes', hue='category_name', alpha=0.6, palette='tab20')
plt.xscale('log')
plt.yscale('log')
plt.title('Views vs Likes by Category')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
